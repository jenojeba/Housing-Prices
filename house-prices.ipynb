```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
file_path = '/kaggle/input/house-prices-advanced-regression-techniques/train.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
features = data[['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']]
target = data['SalePrice']
```

```python
features.fillna(features.mean(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
test_file_path = '/kaggle/input/house-prices-advanced-regression-techniques/test.csv'
test_data = pd.read_csv(test_file_path)
```

```python
test_features = test_data[['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']]
test_features.fillna(test_features.mean(), inplace=True)
```

```python
test_predictions = model.predict(test_features)
```

```python
output = pd.DataFrame({'Id': test_data['Id'], 'SalePrice': test_predictions})
```

```python
output_file_path = '/kaggle/working/predictions.csv'
output.to_csv(output_file_path, index=False)
```
